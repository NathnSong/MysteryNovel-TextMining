```python
from konlpy.tag import Kkma
from konlpy.utils import pprint
kkma = Kkma()
pprint(kkma.sentences(u'네, 안녕하세요. 반갑습니다.')) # 문장 분리
pprint(kkma.nouns(u'질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요!')) # 명사 추출
# 품사 태깅 -> NNG(일반 명사), JX(보조사), SP(구분 기호(쉼표)), MAG(일반 부사), JKB(부사격 조사), JKO(목적격 조사), SF(마침표, 느낌표)
pprint(kkma.pos(u'오류보고는 실행환경, 에러메세지와 함께 설명을 자세히!'))
```

## OKT를 이용해 '그리고 아무도 없었다' 소설 형태소 분석

```python
!pip install wordcloud
```

```python
# 형태소 분석기 Okt 로딩
from konlpy.tag import Okt
okt = Okt()

filename = '그리고 아무도 없었다.txt'
with open(filename, 'r', encoding='utf-8') as f:
  text = f.read()
  # text파일 문단 단위로 자르기
  chunks = text.split('\n\n')
  print(f"총 문단 수 : {len(chunks)}")

  stopwords = [
    '.', ',', '?', '!', '\n', '\n  ', '「', '」', '"', "'",  # 특수문자 및 개행
    '가', '이', '은', '는', '을', '를', '에', '의', '도', '으로', '한',  # 조사 및 의존형태
    '그', '이', '저', '것', '수', '등', '들', '좀', '더', '등등', '하며', '하고', '또',
    '아니', '있', '하', '되', '이렇', '저렇', '같', '때', '듯', '내', '나', '우리', '당신'  # 불명확한 대명사/동사/부사
]

  #nouns
  all_nouns = []
  for chunk in chunks:
    all_nouns.extend(okt.nouns(chunk))

  #morphs
  all_morphs = []
  for chunk in chunks:
    all_morphs.extend(okt.morphs(chunk))

  #phrases
  all_phrases = []
  for chunk in chunks:
    all_phrases.extend(okt.phrases(chunk))

  #pos
  all_pos = []
  for chunk in chunks:
    all_pos.extend(okt.pos(chunk))

# 형태소 분석 결과 출력하기
from collections import Counter

# 불용어 제거하기 & 길이가 1인 경우도 제외
all_nouns = [noun for noun in all_nouns if noun not in stopwords and len(noun) > 1]
all_morphs = [m for m in all_morphs if m not in stopwords and len(m) > 1]

nouns_counter = Counter(all_nouns).most_common(20)
morphs_counter = Counter(all_morphs).most_common(20)

print("명사 top 20:", nouns_counter)
print("형태소 top 20", morphs_counter)
print("구(phrases) 10:", all_phrases[:20])
print("품사태깅 10", all_pos[:20])
```
